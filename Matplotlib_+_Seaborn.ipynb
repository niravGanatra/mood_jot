## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Nice default size
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
# Sample business-like dataset
sales = pd.DataFrame({
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "Product_A": [120,135,110,150,160,170,155,165,172,180,190,210],
    "Product_B": [80,95,90,100,115,120,125,135,140,150,145,155],
    "Region": (["North"]*6)+(["South"]*6)
})
sales["Total"] = sales["Product_A"] + sales["Product_B"]
sales.head()
```

## Matplotlib essentials: Figure, Axes, labels, styles

```python
fig,ax=plt.subplots()
ax.plot(sales["Month"],sales["Product_A"],marker="o") #ax.plot("x-axis","y-axis")
ax.set_title("Sales of Product A")
ax.set_xlabel("Month")
ax.set_ylabel("Sales")
ax.grid(True,alpha=0.3)
```

## Core Matplotlib charts: line, bar, scatter, hist, box

### Multi-line chart + legend

```python
fig,ax=plt.subplots()
ax.plot(sales["Month"],sales["Product_A"],marker="o") #ax.plot("x-axis","y-axis")
ax.plot(sales["Month"],sales["Product_B"],marker="x") #ax.plot("x-axis","y-axis")
ax.set_title("Sales of Product A and B")
ax.set_xlabel("Month")
ax.set_ylabel("Sales")
ax.grid(True,alpha=0.3)
```

### Bar chart (discrete comparison)

```python
fig,ax=plt.subplots()
ax.bar(sales["Month"],sales["Product_A"])
ax.set_title("Sales of Product A")
ax.set_xlabel("Month")
ax.set_ylabel("Sales")
ax.grid(True,alpha=0.3)
```

### Scatter (relationship)

```python
fig,ax=plt.subplots()
ax.scatter(sales["Product_A"],sales["Product_B"])
ax.set_title("Scatter plot of Product A and B")
ax.set_xlabel("Product A")
ax.set_ylabel("Product B")
```
